```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import tensorflow
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras import utils
import numpy as np
model = load_model('google_face.h5')
img = image.load_img('modiji.jpeg', target_size=(64, 64))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
img_data = preprocess_input(x)
y_proba = model.predict(img_data).argmax()
```

```python
y_proba
```

```python
mask = 'With Mask'
without_mask = 'Without Mask'
if (y_proba == 1):
    import cv2
    img_f = plt.imread('modiji.jpeg')
    
    plt.text(100, 100, mask, bbox=dict(facecolor='green', alpha=0.5))
    plt.imshow(img_f)
elif (y_proba==0):
    print(without_mask)
    img_wm = plt.imread('modiji.jpeg')
    plt.text(100,100,without_mask,bbox = dict(facecolor = 'red',alpha = 0.5))
    plt.imshow(img_wm)
```
